Нарисуйте график функции:<br>
y(x) = k∙cos(x – a) + b<br>
для некоторых (2-3 различных) значений параметров k, a, b<br>

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3*np.pi, 3*np.pi, 201)
plt.plot(x, np.cos(x-1))
plt.plot(x, np.cos(2*x-1)+1)
plt.plot(x, np.cos(3*x-1)+4)
plt.plot([-10, 10], [0,0], color='red', linewidth=0.75, linestyle='--')
plt.grid(True)
plt.show()
```
