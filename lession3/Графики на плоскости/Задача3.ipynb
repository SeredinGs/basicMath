### ЗАДАНИЕ
1.Напишите код, который будет переводить полярные координаты в декартовы.

Формулы перевода полярной системы координат:<br>
x = R*cos(alpha)<br>
y = R*sin(alpha)<br>
R = scrt(x^2+y^2)

```python
alpha = np.pi/3
R = 15
x = R * np.cos(alpha)
y = R * np.sin(alpha)
Rt = pow(x**2 + y**2, 0.5)
print(f'Координаты равны ({x, y})')
print(f'Проверка Rt={Rt}')
```

Преобразование выполнено верно

Исходя из формулы полярной системы координат, окружность можно получить когда функция не зависит от alpha.<br>
