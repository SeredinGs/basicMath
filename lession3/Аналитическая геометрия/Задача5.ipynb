Нарисуйте трехмерный график двух параллельных плоскостей.<br>

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X1 = np.arange(-5, 5, 0.5)
Y1 = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X1, Y1)
Z1 = 2 * X + Y
Z2 = 2 * X + Y - 5
ax.plot_wireframe(X, Y, Z1, color='red')
ax.plot_wireframe(X, Y, Z2, color='green')
show()
```

Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
fig = figure()
ax = Axes3D(fig)
X1 = np.arange(-5, 5, 0.5)
Y1 = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X1, Y1)
Z1 = X**2 + Y**2
Z2 = X**2 + (Y - 5)**2
ax.plot_wireframe(X, Y, Z1, color='red')
ax.plot_wireframe(X, Y, Z2, color='blue')
show()
```
