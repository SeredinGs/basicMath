### Задача 3.1
Дополните код Монте-Карло последовательности независимых испытаний расчетом соответствующих вероятностей (через биномиальное распределение) 
и сравните результаты.

```python
k, n = 0, 10000
a = np.random.randint(0, 2, n)
b = np.random.randint(0, 2, n)
c = np.random.randint(0, 2, n)
d = np.random.randint(0, 2, n)
x = a + b + c + d
for i in range(0, n):
    if x[i] == 2:
        k = k + 1
#print(a, b, c, d)
#print(x)
print(k, n, k/n)
```

Формула Бернулли для биноминального распределения выглядит следующим образом:<br>
P(x) = Cml * 1/2^m, где  <br>

```python
m = 4
l = 2
```

Определим число сочетаний:

```python
Ctot = int(factorial(m) / (factorial(l) * factorial(m - l)))
```

Определим вероятность

```python
P = Ctot * 1/pow(2, m)
P
```

Расчетная вероятность примерно равна значению, посчитанному в коде Монте-Карло

### Задача 5
Дополните код расчетом коэффициента корреляции x и y по формуле

```python
n = 100
r = 0.7
x = np.random.rand(n)
y = r*x + (1 - r)*np.random.rand(n)
plt.plot(x, y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

a = (np.sum(x)*np.sum(y) - n*np.sum(x*y))/(np.sum(x)*np.sum(x) - n*np.sum(x*x))
b = (np.sum(y) - a*np.sum(x))/n

A = np.vstack([x, np.ones(len(x))]).T
a1, b1 = np.linalg.lstsq(A, y)[0]
print(a, b)
print(a1, b1)
plt.plot([0, 1], [b, a + b])
plt.show()
```

Рассчитаем коэффициент корреляции:

```python
R = np.sum((x-np.mean(x))*(y-np.mean(y)))/pow((np.sum(pow(x-np.mean(x),2))*np.sum(pow(y-np.mean(y),2))),0.5)
R
```

Сравним полученный коэффициент, с результатом функции corrcoef

```python
Rcontr = np.corrcoef(x,y)
Rcontr[0,1]
```
